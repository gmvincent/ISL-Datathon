```python
import os
import numpy as np
import pandas as pd

from tqdm.notebook import tqdm

from deepface import DeepFace
```

```python
output = pd.DataFrame(columns=["filename", "gender", "race", "age"])
```

```python
folder = "/content/faceimages"
files = os.listdir(folder)
```

```python
for img in tqdm(files, desc="Processing images", total=len(files)):
    try:
        result = DeepFace.analyze(img_path=os.path.join(folder, img),
                                  actions=['age', 'gender', 'race'], silent=True)

        age = result[0]['age']
        gender = result[0]['dominant_gender']
        race = result[0]['dominant_race']

        if np.isnan(age) or age < 0:
            age = -1

        if gender not in ["Man", "Woman"]:
            gender = "unknown"

        if not isinstance(race, str):
            race = "unknown"

        output = output.append(
            {"filename": img, "gender": gender, "race": race, "age": age},
            ignore_index=True)

    except ValueError as e:
        print(f"Error processing {img}: \t {e}")
        output = output.append(
            {"filename": img, "gender": "unknown", "race": "unknown", "age": -1}
            , ignore_index=True)
```

```python
output.to_csv('faceimages.csv', index=False)
```
